###**1.1** This paper is about creating an easier, more standardized process for data cleaning called data tidying. This way it is simple to tidy messy datasets, tidy datasets have a specific structure, and are easy to manipulate.

###**1.2** The “tidy data standard” is intending to make data cleaning a much easier process rather than “start from scratch and reinvent the wheel every time”, as the author puts it. It also works to simplify the data analysis since the tidy datasets and tidy tools work well together.
###**1.3** The first sentence means that making datasets tidy gives a standardized way to compare them and creates a commonality between all tidy datasets, like how all happy families have happiness/love in common. On the other hand, there are many ways in which a dataset could be messy making it harder to compare to others, just how many different factors could cause a family to be unhappy. The second sentence means that for some variables it can be hard to determine if they are meant to be separate variables or if they are both values of a larger variable. For example, as the article talks about, ‘Height’ and ‘Width’ could either be two separate variables or they could be values of a ‘Dimension’ variable.

###**1.4** Wickham defines values as numbers or strings that belong to both a variable and an observation. Variables consist of values that measure the same attribute across all units, while an observation contains all of the values across all attributes for just one unit.

###**1.5** “Tidy Data” is defined as a dataset that has the structure where each variable forms a column, each observation forms a row, and each type of observational unit forms a table.

###**1.6** The 5 most common problems with messy datasets are
1. Column headers are values, not variable names.
2. Multiple variables are stored in one column.
3. Variables are stored in both rows and columns.
4. Multiple types of observational units are stored in the same table.
5. A single observational unit is stored in multiple tables.

The data in Table 4 are messy because there are 6 columns that all have to do with income. Those columns should be melted into one income column, and there should be an additional column for the frequency.  Melting a dataset is where you “stack it” or turn columns into rows.

###**1.7** Table 11 is messy because there are 31 day columns that each represent a different day of the month and these need to be combined to be considered tidy. Table 12 is tidy because it instead has one date column which includes all of those previous day columns, and also has one column that includes all of the values. Now, each variable has one column.

###**1.8** The “chicken-and-egg” problem with tidy data is that if tidy data is only as useful as the tidy tools created for tidy data, then the tools and the data will be too highly linked together and the tools won’t work for untidy data. Wickham hopes in the future that others will build upon this framework to better develop more tools and better ones as well. He also hopes that other frameworks can be developed to make other aspects of data cleaning easier as well such as identifying missing values and parsing dates and numbers.

### **<small> Explain the choices you make. </small>**
After looking at the unique values and their value counts in the 'Type' column of the sharks data, I determined that I needed to combine many columns.
```python
# 2.3
import pandas as pd
import numpy as np
pretrial_data = pd.read_csv('/content/VirginiaPretrialData2017.csv')
```

```python

#rename column
pretrial_data = pretrial_data.rename(columns={'WhetherDefendantWasReleasedPretrial': 'ReleasedPretrial'})
#pretrial_data.head()
print(pretrial_data['ReleasedPretrial'].value_counts())
#9 = "unclear" so we should turn these into NaN
pretrial_data['ReleasedPretrial'] = pretrial_data['ReleasedPretrial'].replace(9,np.nan)
print(pretrial_data['ReleasedPretrial'].value_counts())

nan_count = pretrial_data['ReleasedPretrial'].isna().sum() #check to see number of NaNs per column
print(nan_count)
```

### **<small> Explain the choices you make. </small>**
I renamed the 'WhetherDefendantWasReleasedPretrial' column to 'ReleasedPretrial'and also replaced any 9s with np NaNs since 9 in the codebook was "unclear".

```python
# 2.4
#rename columns
pretrial_data = pretrial_data.rename(columns={'ImposedSentenceAllChargeInContactEvent': 'Sentence'})
pretrial_data = pretrial_data.rename(columns={'SentenceTypeAllChargesAtConvictionInContactEvent': 'SentenceType'})
print(pretrial_data['SentenceType'].value_counts())
# 9 is "Not applicable", 4 is "pending,dismissed...", 1 is "probation/no incarceration"

#replace certain values of Sentence with np.nan depending on values of SentenceType
pretrial_data.loc[ pretrial_data['SentenceType'] == 9, 'Sentence'] = np.nan #record not found or unknown
pretrial_data.loc[ pretrial_data['SentenceType'] == 4, 'Sentence'] = 0   #charges pending, dismissed, or deferred
pretrial_data.loc[ pretrial_data['SentenceType'] == 0, 'Sentence'] = 0   #Probation/no incarceration

#crosstab
pretrial_data['sentence_NA'] = pretrial_data['Sentence'].isnull()
print(pd.crosstab(pretrial_data['sentence_NA'] ,pretrial_data['SentenceType']),'\n')

```

### **<small> Explain the choices you make. </small>**
After renaming the columns to 'Sentence' and 'SentenceType', I looked at the codebook to see what the different values of 'SentenceType' meant. I decided I wanted to replace certain values of Sentence with np.nan for the value of SentenceType 9 since 9 was "Not applicable (e.g., disposition record not found, or unknown)". For SentenceType values 0 and 4, I decided to replace the NA Sentence value with 0 since 0 meant "probation/no incarceration" and 4 was "pending,dismissed..." so these resulted in a sentence length of 0.
Then I used crosstab to view these changes, now Sentence is only missing where SentenceType is 9 (no record found).
